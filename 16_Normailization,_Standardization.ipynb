```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample data from a Gaussian distribution
np.random.seed(42)
# Generate data from a standard normal distribution (mean 0, variance 1)
data = np.random.randn(100, 1)
# Scale and shift to a desired mean and standard deviation (e.g., mean 50, std dev 20)
mean = 50
std_dev = 20
data = data * std_dev + mean


# Create a DataFrame
df = pd.DataFrame(data, columns=['Original Data'])

# Apply Min-Max Normalization
scaler_minmax = MinMaxScaler()
df['Normalized Data (Min-Max)'] = scaler_minmax.fit_transform(df[['Original Data']])

# Apply Standardization
scaler_standard = StandardScaler()
df['Standardized Data'] = scaler_standard.fit_transform(df[['Original Data']])

# Display the first few rows
display(df.head())
```

```python
# Calculate mean and variance for each column
mean_original = df['Original Data'].mean()
variance_original = df['Original Data'].var()

mean_normalized = df['Normalized Data (Min-Max)'].mean()
variance_normalized = df['Normalized Data (Min-Max)'].var()

mean_standardized = df['Standardized Data'].mean()
variance_standardized = df['Standardized Data'].var()

# Display the results
print(f"Original Data - Mean: {mean_original:.4f}, Variance: {variance_original:.4f}")
print(f"Normalized Data (Min-Max) - Mean: {mean_normalized:.4f}, Variance: {variance_normalized:.4f}")
print(f"Standardized Data - Mean: {mean_standardized:.4f}, Variance: {variance_standardized:.4f}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original data
axes[0].hist(df['Original Data'], bins=20, color='skyblue')
axes[0].set_title('Original Data')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Plot normalized data
axes[1].hist(df['Normalized Data (Min-Max)'], bins=20, color='lightcoral')
axes[1].set_title('Normalized Data (Min-Max)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

# Plot standardized data
axes[2].hist(df['Standardized Data'], bins=20, color='lightgreen')
axes[2].set_title('Standardized Data')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Generate sample data with outliers
np.random.seed(42)
data_with_outliers = np.random.rand(100, 1) * 100
outliers = np.random.rand(10, 1) * 1000 + 200 # Add some outliers
data_with_outliers = np.vstack((data_with_outliers, outliers))

# Scale and shift to a desired mean and standard deviation (e.g., mean 50, std dev 20)
mean = 50
std_dev = 20
data_with_outliers = data_with_outliers * std_dev + mean


# Create a DataFrame
df_outliers = pd.DataFrame(data_with_outliers, columns=['Original Data with Outliers'])

# Apply Min-Max Normalization
scaler_minmax_outliers = MinMaxScaler()
df_outliers['Normalized Data (Min-Max) with Outliers'] = scaler_minmax_outliers.fit_transform(df_outliers[['Original Data with Outliers']])

# Apply Standardization
scaler_standard_outliers = StandardScaler()
df_outliers['Standardized Data with Outliers'] = scaler_standard_outliers.fit_transform(df_outliers[['Original Data with Outliers']])

# Display the first few rows
display(df_outliers.head())
display(df_outliers.tail())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original data with outliers
axes[0].hist(df_outliers['Original Data with Outliers'], bins=20, color='skyblue')
axes[0].set_title('Original Data with Outliers')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Plot normalized data with outliers
axes[1].hist(df_outliers['Normalized Data (Min-Max) with Outliers'], bins=20, color='lightcoral')
axes[1].set_title('Normalized Data (Min-Max) with Outliers')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

# Plot standardized data with outliers
axes[2].hist(df_outliers['Standardized Data with Outliers'], bins=20, color='lightgreen')
axes[2].set_title('Standardized Data with Outliers')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample 2D data
np.random.seed(42)
data_2d = np.random.rand(100, 2) * 100

# Create a DataFrame
df_2d = pd.DataFrame(data_2d, columns=['Feature 1', 'Feature 2'])

# Apply Min-Max Normalization
scaler_minmax_2d = MinMaxScaler()
df_2d_normalized = pd.DataFrame(scaler_minmax_2d.fit_transform(df_2d), columns=['Feature 1 Normalized', 'Feature 2 Normalized'])

# Apply Standardization
scaler_standard_2d = StandardScaler()
df_2d_standardized = pd.DataFrame(scaler_standard_2d.fit_transform(df_2d), columns=['Feature 1 Standardized', 'Feature 2 Standardized'])

# Combine dataframes for easier plotting
df_2d_combined = pd.concat([df_2d, df_2d_normalized, df_2d_standardized], axis=1)

display(df_2d_combined.head())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot original 2D data
axes[0].scatter(df_2d_combined['Feature 1'], df_2d_combined['Feature 2'], color='skyblue')
axes[0].set_title('Original 2D Data')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')
axes[0].set_aspect('equal', adjustable='box') # Set aspect ratio to equal

# Plot normalized 2D data
axes[1].scatter(df_2d_combined['Feature 1 Normalized'], df_2d_combined['Feature 2 Normalized'], color='lightcoral')
axes[1].set_title('Normalized 2D Data (Min-Max)')
axes[1].set_xlabel('Feature 1 Normalized')
axes[1].set_ylabel('Feature 2 Normalized')
axes[1].set_aspect('equal', adjustable='box') # Set aspect ratio to equal
axes[1].set_xlim([-0.1, 1.1]) # Add limits to better visualize the 0-1 range
axes[1].set_ylim([-0.1, 1.1])

# Plot standardized 2D data
axes[2].scatter(df_2d_combined['Feature 1 Standardized'], df_2d_combined['Feature 2 Standardized'], color='lightgreen')
axes[2].set_title('Standardized 2D Data')
axes[2].set_xlabel('Feature 1 Standardized')
axes[2].set_ylabel('Feature 2 Standardized')
axes[2].set_aspect('equal', adjustable='box') # Set aspect ratio to equal

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample 2D data with different scales
np.random.seed(42)
data_2d_scaled = np.random.rand(100, 2)
data_2d_scaled[:, 0] = data_2d_scaled[:, 0] * 1000  # Make the first feature have a larger scale
data_2d_scaled[:, 1] = data_2d_scaled[:, 1] * 10    # Keep the second feature with a smaller scale

# Create a DataFrame
df_2d_scaled = pd.DataFrame(data_2d_scaled, columns=['Large Scale Feature', 'Small Scale Feature'])

# Apply Min-Max Normalization
scaler_minmax_2d_scaled = MinMaxScaler()
df_2d_scaled_normalized = pd.DataFrame(scaler_minmax_2d_scaled.fit_transform(df_2d_scaled), columns=['Large Scale Feature Normalized', 'Small Scale Feature Normalized'])

# Apply Standardization
scaler_standard_2d_scaled = StandardScaler()
df_2d_scaled_standardized = pd.DataFrame(scaler_standard_2d_scaled.fit_transform(df_2d_scaled), columns=['Large Scale Feature Standardized', 'Small Scale Feature Standardized'])

# Combine dataframes for easier plotting
df_2d_scaled_combined = pd.concat([df_2d_scaled, df_2d_scaled_normalized, df_2d_scaled_standardized], axis=1)

display(df_2d_scaled_combined.head())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot original 2D data with different scales
axes[0].scatter(df_2d_scaled_combined['Large Scale Feature'], df_2d_scaled_combined['Small Scale Feature'], color='skyblue')
axes[0].set_title('Original 2D Data with Different Scales')
axes[0].set_xlabel('Large Scale Feature')
axes[0].set_ylabel('Small Scale Feature')
# axes[0].set_aspect('equal', adjustable='box') # Removing equal aspect ratio to better show the scale difference

# Plot normalized 2D data with different scales
axes[1].scatter(df_2d_scaled_combined['Large Scale Feature Normalized'], df_2d_scaled_combined['Small Scale Feature Normalized'], color='lightcoral')
axes[1].set_title('Normalized 2D Data (Min-Max) with Different Scales')
axes[1].set_xlabel('Large Scale Feature Normalized')
axes[1].set_ylabel('Small Scale Feature Normalized')
axes[1].set_aspect('equal', adjustable='box') # Set aspect ratio to equal for comparison after scaling
axes[1].set_xlim([-0.1, 1.1]) # Add limits to better visualize the 0-1 range
axes[1].set_ylim([-0.1, 1.1])

# Plot standardized 2D data with different scales
axes[2].scatter(df_2d_scaled_combined['Large Scale Feature Standardized'], df_2d_scaled_combined['Small Scale Feature Standardized'], color='lightgreen')
axes[2].set_title('Standardized 2D Data with Different Scales')
axes[2].set_xlabel('Large Scale Feature Standardized')
axes[2].set_ylabel('Small Scale Feature Standardized')
axes[2].set_aspect('equal', adjustable='box') # Set aspect ratio to equal for comparison after scaling


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample 2D data with different scales and outliers
np.random.seed(42)
data_2d_scaled_outliers = np.random.rand(100, 2)
data_2d_scaled_outliers[:, 0] = data_2d_scaled_outliers[:, 0] * 1000  # Large scale feature
data_2d_scaled_outliers[:, 1] = data_2d_scaled_outliers[:, 1] * 10    # Small scale feature

# Add outliers to both features
outliers_2d = np.random.rand(10, 2)
outliers_2d[:, 0] = outliers_2d[:, 0] * 5000 + 2000 # Outliers for large scale feature
outliers_2d[:, 1] = outliers_2d[:, 1] * 50 + 20   # Outliers for small scale feature

data_2d_scaled_outliers = np.vstack((data_2d_scaled_outliers, outliers_2d))

# Create a DataFrame
df_2d_scaled_outliers = pd.DataFrame(data_2d_scaled_outliers, columns=['Large Scale Feature with Outliers', 'Small Scale Feature with Outliers'])

# Apply Min-Max Normalization
scaler_minmax_2d_scaled_outliers = MinMaxScaler()
df_2d_scaled_outliers_normalized = pd.DataFrame(scaler_minmax_2d_scaled_outliers.fit_transform(df_2d_scaled_outliers), columns=['Large Scale Feature Normalized with Outliers', 'Small Scale Feature Normalized with Outliers'])

# Apply Standardization
scaler_standard_2d_scaled_outliers = StandardScaler()
df_2d_scaled_outliers_standardized = pd.DataFrame(scaler_standard_2d_scaled_outliers.fit_transform(df_2d_scaled_outliers), columns=['Large Scale Feature Standardized with Outliers', 'Small Scale Feature Standardized with Outliers'])

# Combine dataframes for easier plotting
df_2d_scaled_outliers_combined = pd.concat([df_2d_scaled_outliers, df_2d_scaled_outliers_normalized, df_2d_scaled_outliers_standardized], axis=1)

display(df_2d_scaled_outliers_combined.head())
display(df_2d_scaled_outliers_combined.tail())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot original 2D data with different scales and outliers
axes[0].scatter(df_2d_scaled_outliers_combined['Large Scale Feature with Outliers'], df_2d_scaled_outliers_combined['Small Scale Feature with Outliers'], color='skyblue')
axes[0].set_title('Original 2D Data with Different Scales and Outliers')
axes[0].set_xlabel('Large Scale Feature with Outliers')
axes[0].set_ylabel('Small Scale Feature with Outliers')


# Plot normalized 2D data with different scales and outliers
axes[1].scatter(df_2d_scaled_outliers_combined['Large Scale Feature Normalized with Outliers'], df_2d_scaled_outliers_combined['Small Scale Feature Normalized with Outliers'], color='lightcoral')
axes[1].set_title('Normalized 2D Data (Min-Max) with Different Scales and Outliers')
axes[1].set_xlabel('Large Scale Feature Normalized with Outliers')
axes[1].set_ylabel('Small Scale Feature Normalized with Outliers')
axes[1].set_aspect('equal', adjustable='box') # Set aspect ratio to equal for comparison after scaling
axes[1].set_xlim([-0.1, 1.1]) # Add limits to better visualize the 0-1 range
axes[1].set_ylim([-0.1, 1.1])


# Plot standardized 2D data with different scales and outliers
axes[2].scatter(df_2d_scaled_outliers_combined['Large Scale Feature Standardized with Outliers'], df_2d_scaled_outliers_combined['Small Scale Feature Standardized with Outliers'], color='lightgreen')
axes[2].set_title('Standardized 2D Data with Different Scales and Outliers')
axes[2].set_xlabel('Large Scale Feature Standardized with Outliers')
axes[2].set_ylabel('Small Scale Feature Standardized with Outliers')
axes[2].set_aspect('equal', adjustable='box') # Set aspect ratio to equal for comparison after scaling

plt.tight_layout()
plt.show()
```
